```python
losses = np.genfromtxt('/home/fan/work/python/results/sadHybridHumanPredictor/ensemble_losslog.csv', delimiter=',')
```

```python
losses.shape
```

```python
losses = losses[:85 * 9, :-1:5]
```

```python
import pandas as pd
labels = ['ensemble']
for i in xrange(31):
    labels.append('model {}'.format(i + 1))
```

```python
losses_df = pd.DataFrame(losses)
```

```python
sns.plt.clf()
sns.set_palette(sns.color_palette('cubehelix', 32))
sns.plt.figure(figsize=(40, 7))
sns.plt.plot(losses_df)
sns.plt.legend(labels=labels, bbox_to_anchor=(0.5, -0.2), loc='lower center', ncol=16, borderaxespad=1, fontsize=16)
sns.plt.show()
```

```python
target_days = [0, 1, 2, 3, 4, 7]
losses_array = [losses[i * 85: (i + 1) * 85] for i in target_days]
```

```python
sns.plt.clf()
sns.set_palette([(0, 0, 0)] + sns.color_palette('hls', 31))
fig, axes = sns.plt.subplots(2, 3, figsize=(20, 10))
sns.plt.subplots_adjust(hspace=0.32)
titles = map(lambda x:'Jan {}'.format(x + 1), target_days)
for i in xrange(6):
    losses_df = pd.DataFrame(losses_array[i])
    row = i // 3
    col = i % 3
    ax_curr = axes[row, col]
    ax_curr.plot(losses_df)
    ax_curr.set_xlabel('Time')
    ax_curr.set_ylabel('Cross Entropy')
    ax_curr.set_title(titles[i])
sns.plt.legend(labels=labels, bbox_to_anchor=(1, 2.5), ncol=1, borderaxespad=1.8, fontsize=12)
sns.plt.show()
```
